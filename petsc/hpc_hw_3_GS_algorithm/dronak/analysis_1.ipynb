```python
vals = {1000:1.3110e-02, 5000:2.0242e-02, 10000:3.0296e-02, 50000:9.7507e-02, 100000:1.8525e-01}
```

```python
# Does the time T of your runs above have a polynomial dependence on the vector size N, T=CN^alpha, where C is some constant? If so, what is alpha?
# alpha can be found by taking the log of both sides of the equation and solving for alpha.  The slope of the line is alpha. (log(T) = log(C) + alpha*log(N))
alpha = np.log(1.3110e-02) - np.log(1.8525e-01)
alpha = alpha / (np.log(1000) - np.log(100000))
print("alpha = ", alpha)

# solving for C: log(T) = log(C) + alpha*log(N)
C = 1.3110e-02 / (1000**alpha)
print("C = ", C)

# plot the line
x = np.linspace(1000, 100000, 100)
y = C * x**alpha
plt.plot(x, y, 'r-', label='T = CN^alpha')
plt.plot(vals.keys(), vals.values(), 'o-', label='ground truth')
plt.xlabel('N')
plt.ylabel('Time (s)')
plt.legend()
plt.title('Problem 2: Runtime for fixed vector set k=20 with varying N')
plt.show()
```
